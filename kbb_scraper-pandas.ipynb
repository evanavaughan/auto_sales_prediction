```python
def clean(url):
    url = url.replace('staging.', '')
    url = url.replace('//', '//www.')
    seperator = '?'
    return url.split(seperator, 1)[0]
```

```python
prefix = 'https://www.kbb.com/car-finder/'
suffix = '/?categories=sedan&years=2006-2020'
```

```python
def scrape(pre, suf):
    urls = []
    for i in range(1,139):
        site  = pre + str(i) + suf
        html = requests.get(site).text
        soup = BeautifulSoup(html, 'lxml')
        for link in soup.find_all('a'):
            urls.append(link.get('href'))
    cleaned_list = [clean(url) for url in urls if url.startswith('https://staging')]
    return cleaned_list
```

```python
listy = []
for u in cleaned_list:
    u = u.replace('kbb', 'www.kbb')
    listy.append(u)
```

```python
def write_list_to_file(guest_list, filename):
    """Write the list to csv file."""

    with open(filename, "w") as outfile:
        for entries in guest_list:
            outfile.write(entries)
            outfile.write("\n")
```

```python
def combined_mpg(souper):
    combined = souper.find('p', {'class':'label-two'}).text
    combined = int(combined.split()[0])
    return combined
```

```python
def city_hwy_mpg(souper):
    mpg = souper.find('p', {'class':'label-three'}).text.split()
    city = int(mpg[0])
    hwy = int(mpg[3])
    return city, hwy
```

```python
def ratings(souper):
    rating = souper.find_all('p', {'class':'rating-text'})
    expert_rating = float(rating[0].text.split()[0])
    user_rating = float(rating[1].text.split()[0])
    return expert_rating, user_rating
```

```python
def price_low_high(souper):
    price = souper.find_all('p', {'class':'label-two range'})[0].text
    price = price.replace('$', '')
    price = price.replace(',', '')
    price = price.split()
    low = int(price[0])
    high = int(price[2])
    return low, high
```

```python
def read_site_html(url):
    '''
    pulls text of tables from inputted url
    returns as list of dataframes
    '''
    html = requests.get(url).text
    dataframes = pd.read_html(html, header=0)
    return dataframes
```
